```
from itertools import permutations
```

```
perm = list(permutations([1,2,3,4,5,6]))
```

There are $6!$ different possible permutations of these.  This is 
$$
6! = 6\times 5\times 4\times 3 \times 3\times 2\times 1 = 720
$$

```
for i in perm:
  print(i)
```

```
6*0.37
```

So the theory suggests that we interview two people and then take the one we see that is better than all the rest.

```
import random 
example = perm[random.randrange(720)]
print(example)
```

```
def nextbest(permutation, lookbest, startlookingat):
  for indiv in permutation[startlookingat:]:
    if indiv > lookbest:
      return indiv
  return 0 



nextbest(example, 5,2)

```

```
numberOfBests = 0
for i in perm:
  maximum = max(i[0:2])
  if nextbest(i,maximum,2) == 6:
    numberOfBests = numberOfBests + 1
```
