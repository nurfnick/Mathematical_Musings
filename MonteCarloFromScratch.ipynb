# Random Numbers, Monte Carlo and Las Vegas Methods


```python
a = 2**16
m = 2**32-1
c = 578492301263467
x = 740202648000

xlist = []
```

```python
import math
math.gcd(a,m)
```

`a` and `m` need to be relatively prime for this scheme to work

```python
xlist.append(x)

for i in range(10000000):
  x = (a*x+c)%m
  xlist.append(x)
```

```python
xlist
```

To get pseudo - random between $[0,1)$, we divide each entry by `m` 

```python
xlist1 =[]

for i in xlist:
  xlist1.append(i/m)
```

```python
xlist1
```

Now for the Monte Carlo simulation.  Calculating the area of the first quadrant of the circle.

```python
N = 0
M = 500000

for i in range(M):
  u,v = xlist1[2*i],xlist1[2*i+1]
  if (u**2+v**2<1):
    N+=1
print(N/M)
```
