# Crazy Derivatives

I saw the following function in this meme.

image.png

and was curious what it looked like to not have a convergent derivative.  I numerically investigate this below.

```python
import math
import numpy as np

def f(x,n):
  sum = 0
  for i in range(n):
    sum += np.cos(13**i*math.pi*x)/2**i
  return sum
```

```python
import matplotlib.pyplot as plt


x = np.linspace(0,1,10000)
plt.plot(x,f(x,60))
```

```python
def der(function,x,stepsize):
  return (function(x + stepsize)-function(x))/stepsize
```

```python
fun = lambda x:f(x,25)
```

```python
der(fun,0,.00000000001)
```

```python
l = []
for i in range(1,50):
  l.append(10**(-1*i))
```

```python
for item in l:
  print(item,der(fun,0,item))
```

A note on why the function makes sense but the derivative does not:

$$ f(x) =\sum_{n=0}^\infty\frac{\cos(13^n\pi x)}{2^n}$$

This function is convergent by the squeeze theorem:  $-1\leq \cos(13^n\pi x)\leq 1$

But the derivative:

$$ f'(x) = \sum_{n=0}^\infty \frac{-13^n\sin(13^n\pi x)}{2^n} $$

will not coverge!
